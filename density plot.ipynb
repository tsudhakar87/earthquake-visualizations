```python
# import
import matplotlib.pyplot as plt
import pandas as pd
import altair as alt
import geopandas as gpd
import geodatasets
from shapely.geometry import Point
import zipfile
import os
import folium
```

```python
# load dataset
earthquakes = pd.read_csv('earthquake_merged2.csv')
eq = earthquakes.dropna(subset=['Latitude', 'Longitude'])
```

```python
import seaborn as sns
sns.kdeplot(eq['Mag'], bw_adjust = 1, fill = True)
plt.xlabel('Magnitude')
plt.ylabel('Density')
plt.title('Magnitude Distribution in Earthquakes since 1900')
plt.show()
```

The graph above represents density plot of the distribution of earthquake magnitudes, which we used to explore the severity of earthquakes. The graph shows that earthquake magnitudes has a slightly bimodal distribution, with peaks around 6.0 and 8.0 in magnitude. In addition, the distribution is relatively narrow, with very little earthquakes being exceptionally low or high in magnitude, so we can deduce that the kurtosis is quite high.
